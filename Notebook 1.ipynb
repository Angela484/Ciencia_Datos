### 4. Limpieza del df

```
#funcion para hacer limpieza en columnas específicas, string
def Limpieza_String (df,column_name):
    df[column_name] = df[column_name].str.strip() ##Limpieza de datos en blanco
    df[column_name] = df[column_name].str.upper() #colocar todo en mayúscula
    df[column_name] = df[column_name].replace("  ","") #limpieza de dobles espacios
    df[column_name] = df[column_name].replace('[^\\w\\d\\s\\.]','',regex=True) #Expresiones regulares para quitar caracteres especiales
    df[column_name] = df[column_name].replace('$','') #reemplazo del signo $$
    df[column_name] = df[column_name].str.replace('Á', 'A') ## limpieza de las tildes 
    df[column_name] = df[column_name].str.replace('É', 'E')
    df[column_name] = df[column_name].str.replace('Í', 'I')
    df[column_name] = df[column_name].str.replace('Ó', 'O')
    df[column_name] = df[column_name].str.replace('Ú', 'U')
    return df
```

```
#funcion para hacer limpieza en columnas específicas, numéricas
def Limpieza_Numericas (df,column_name):
    df[column_name] = df[column_name].replace('[^0-9]','',regex=True) #quitar todo lo diferente a números
    df[column_name] = df[column_name].replace('','0').astype(float) # colocarle un cero a las variables vacías, con el float, cambio el formato a la variable. 
    return df
```

```
#limpieza de variables numéricas y transformación del tipo de dato.
Limpieza_Numericas (df_1,"nit_entidad")
Limpieza_Numericas (df_1,"telefono_entidad")
Limpieza_Numericas (df_1,"ppto_total")
Limpieza_Numericas (df_1,"limite_menor_cuantia")
Limpieza_Numericas (df_1,"limite_minima_cuantia")
Limpieza_Numericas (df_2,'nit_entidad')
Limpieza_Numericas (df_2,'telefono_contacto')
Limpieza_Numericas (df_2,'valor_presupuesto_general')
Limpieza_Numericas (df_2,'presupuesto_menor_cuantia')
Limpieza_Numericas (df_2,'presupuesto_minima_cuantia')
```
